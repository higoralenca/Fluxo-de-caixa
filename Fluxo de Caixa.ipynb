```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tabela_caixa = pd.read_excel("Fluxo de Caixa.xlsm", sheet_name="Fluxo de Caixa")
pd.set_option('display.max_columns', None) # Mostrar todas as colunas
pd.set_option('display.max_rows', None) # Mostrar todas as linhas
display(tabela_caixa)
```

```python
# Movimentação de entradas e saídas


fig = sns.lineplot(tabela_caixa, x="Data", y="Valor", hue="Movimentação")
sns.set(rc={"figure.figsize":(10,5)}) 
plt.xticks(rotation=45)
sns.move_legend(fig, "upper left", bbox_to_anchor=(1, 1)) 
```

```python
# Análise do valor acumulado


sns.lineplot(tabela_caixa, x="Data", y="Acumulado", color="c")
```

```python
# Análise das classificações de movimentação 

sns.jointplot(tabela_caixa, x="Data", y="Classificação", color="#4CB391")
plt.xticks(rotation=45)
```

```python
# Movimentação das contas 

sns.violinplot(tabela_caixa, x="Conta", y="Acumulado")
```

```python
# Finanças dos centros de custos

sns.barplot(tabela_caixa, x="Centro de Custos", y="Acumulado")
sns.set(rc={"figure.figsize":(10,5)})
```

- Movimentação dos produtos

```python
fig=sns.displot(tabela_caixa, x="Data", col="Descrição", hue="Movimentação", kind="ecdf", aspect=.75, linewidth=2, palette="husl")
plt.xticks(rotation=45)
```

```python
sns.histplot(tabela_caixa, x="Data", hue="Descrição")
```
